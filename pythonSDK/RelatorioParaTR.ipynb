```python
import json 
import re
def load_json_file(path: str) -> dict:
    with open(path, 'r') as f:
        return json.load(f)
    
```

```python
items = load_json_file('json1123/itens.json')
volumes = load_json_file('json1123/volumes.json')
```

```python
def filter_itens_by_description(regex: str) -> list:
    return filter(lambda item: re.match(regex, item['short_descricao']), items.values())

def filter_itens_by_function(filter_func: callable ) -> list:
    return filter(filter_func, items.values())
```

```python
for item in filter_itens_by_description(r"CONDICIONAD.*"):
    print(item["key"], item["short_descricao"])
```
